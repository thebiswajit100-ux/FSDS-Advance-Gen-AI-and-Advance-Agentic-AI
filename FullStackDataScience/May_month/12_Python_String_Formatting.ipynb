% Operator

```python
name = 'Pete'
'Hello %s' % name # %s = string placeholder
```

```python
num = 5
'I have %d apples' % num
```

str.format 

```python
# str.format() method: modern string formatting
name = 'John'
age = 20

"Hello i'm {}, my age is {}".format(name, age) # {} = placeholder
```

Formatted String Literals or f-strings

```python
# f-string: recommended way to format stirngs
name = 'Elizabeth'
f'Hello {name}!' # f prefix allows expressions in {}
```

```python
# f-strings support expressions: can include calculations inside {}
a = 5
b = 10
f'Five plus ten is {a + b} and not {2 * (a + b)}.'# Evaluates expressions
```

Multiline f-strings

```python
name = 'Robert'
message = 12
(
    f'Hi, {name}.'
    f'You have {message} unread message'
)
```

The = specifier

```python
# = specifier: prints both variable name and value 
from datetime import datetime
now = datetime.now().strftime("%b/%d/%Y - %H:%M:%S")
f'date and time: {now=}' # prints "now = 'Nov/14/2022 - 20:50:01'"
```

Adding spaces or characters

```python
name = 'Robert'
f"{name.upper() = :-^20}"
```

```python
f"{name.upper() = :^20}"
```

```python
f"{name.upper() = :20}"
```

Formatting Digits

```python
a = 10000000
f"{a:,}"
```

```python
a = 3.1415926
f"{a:2f}"
```

```python
a = 0.816562
f"{a:.2%}"
```

```python
a = 3.1415926
f"{a:.2f}"   # format float 2 decimal places
```

```python
a
```

```python
f"{a:+.2f}" # Format float 2 decimal places with sign
```

```python
a = -1
```

```python
f"{a:+.2f}" # Format float 2 decimal places with sign
```

```python
a = 2.71828 
f"{a:.0f}"  # Format float  with no decimal places
```

```python
a = 4 
f"{a:0>2d}"  # pad number with zeros (left padding, width 2)
```

```python
f"{a:x<4d}"  # pad number with x's (right padding, width 4)
```

```python
a = 10
f"{a:x<4d}"  #pad number with x's (right padding, width 4)
```

```python
a = 1000000
f"{a:,}"     # Number format with comma separator
```

```python
a = 0.35 
f"{a:.2%}"   # Format percentage
```

```python
a = 1000000000
f"{a:.2e}"   # Exponent notation
```

```python
a = 11
f"{a:11d}"   # Right-aligned (default, width 10)
```

```python
f"{a:<11d}"  # Left-aligned (width 10)
```

```python
f"{a:^11d}"   # center aligned (width 10)
```

Template Strings

```python
from string import Template
name = 'Elizabeth'
t = Template('Hey $name!')
t.substitute(name=name)
```

Regular Expressions

```python
# Import re module for regular expression operations
import re

# re.compile(): create regex pattern object (use raw string r'' to avoid escaping)
phone_num_regex = re.compile(r'\d\d\d-\d\d\d-\d\d\d\d')  # Pattern: 3 digits-3 digits-4 digits

mo = phone_num_regex.search('My number is 415-555-4242.') # search for pattern

print(f'Phone number found: {mo.group()}') # group() returns matched text
```
