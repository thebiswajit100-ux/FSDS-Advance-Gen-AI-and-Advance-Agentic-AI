Packing and Unpacking Tuples

```python
# a trailing comma
a = 1 # a is the value 1
a = 1, # a is the tuple (1,)
type(a)
```

```python
a = (1,) # a is the tuple (1,)
print(type(a))
a = (1) # a is the value 1 and not a tuple
print(type(a))
```

```python
# To unpack values from a tuple and do multiple assignments use
x, y, z = (1, 2, 3)
print(x, y, z)
```

```python
a = (1, 2, 3, 4)
_, x, y, _ = a
print(x)
print(y)
```

```python
first, *more, last = (1, 2, 3, 4, 5)
# first == 1
# more == [2, 3, 4]
# last == 5
```

```python
print(first)
print(more)
print(last)
```

Operations on Sets

with other sets

```python
# Intersection
{1, 2, 3, 4, 5}.intersection({3, 4, 5, 6})  # {3, 4, 5}
```

```python
{1, 2, 3, 4, 5} & {3, 4, 5, 6}
```

```python
# Union
{1, 2, 3, 4, 5}.union({3, 4, 5, 6})
```

```python
{1, 2, 3, 4, 5} | {3, 4, 5, 6}
```

```python
# Difference
{1, 2, 3, 4, 5}.difference({2,3,5})
```

```python
{1, 2, 3, 4, 5} - {2, 3, 5}  # {1,4}
```

```python
# Symmetric difference with
{1, 2, 3, 4}.symmetric_difference({2, 3, 5})  # {1, 4, 5}
```

```python
{1, 2, 3, 4} ^ {2, 3, 5}
```

```python
# Superset check
{1, 2}.issuperset({1, 2, 3})  # False
```

```python
{1, 2} >= {1, 2, 3} 
```

```python
# subset check
{1, 2}.issubset({1,2,3})  # True
```

```python
{1, 2} <= {1, 2, 3}
```

```python
# Disjoint check
{1, 2}.isdisjoint({3, 4})  # True
```

```python
{1, 2}.isdisjoint({1, 4})   # False
```

Get the unique elements of a list

```python
restaurants = ["McDonald's", "Burger King", "McDonald's", "Chicken Chicken"]
unique_restaurants = set(restaurants)
```

```python
print(unique_restaurants) # The set is not in the same order as the original list; that is becase
                          # sets are unordered, just like dicts
```

```python
# This can easily transformed back into the list with pythons"built in list function, giving another list
# that si the same list as the original but without duplicates:
list(set(unique_restaurants))
```
