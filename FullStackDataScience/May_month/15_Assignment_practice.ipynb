Set Operation

Union

```python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}
C = {8, 9, 10}
```

```python
A.union(B)
```

```python
A | B # Union of A and B (All Elements from both sets. No Duplicated)
```

```python
A.union(B, C) # Union of A, B and C.
```

```python
"""
Updates the set calling the update() method with union of A , B & C.
For Below example Set A will be updated with union of A, B & C.
"""
A.update(B,C)
A
```

Intersection

```python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}
```

```python
A & B # Intersection of A and B (commn items in both sets)
```

```python
A.intersection(B) 
```

```python
"""
Updates the set calling the intersection_update() method with the intersecion 
For below example set A will be updated with the intersecion of A & B.
"""
A.intersection_update(B)
A
```

Difference

```python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}
```

```python
A - B # set of elements that are only in A but not in B
```

```python
A.difference(B)  # Difference of sets
```

```python
B - A # set of elements that are only in B but not in A
```

```python
B.difference(A)
```

```python
"""
Updates the set calling the difference_update() method with the difference of the set
For below example set B wll be updated with the difference of B & A.
"""
B.difference_update(A)
B
```

Symmetric Difference

```python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}
```

```python
A ^ B  # Symmetric difference (set of elements in A and B but not in both.
```

```python
A.symmetric_difference(B)  # symmetric difference of sets
```

```python
"""
Updates the set calling the symmetric_difference_update() method 
for below example set A will be update with the symmetric difference of A & B.
"""
A.symmetric_difference_update(B)
A
```
