```python
import numpy as np
import pandas as pd
```

```python
# import Matplotlib
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
x1 = np.linspace(0, 10 ,50)
print(x1)
```

```python
# Create a plot figure
# fig = plt.figure()

plt.plot(x1, np.sin(x1), '-')
plt.plot(x1, np.cos(x1), '--')
plt.show()
```

```python
# The following code produces sine and cosine curves using pyplot API.

plt.subplot(2, 1, 1) # rows , columns , panel number

plt.plot(x1, np.cos(x1), '*')

plt.show()
```

```python
# create a plot figure

plt.figure()

plt.subplot(2, 1, 1)  # (rows, columns, panel number)

plt.plot(x1, np.sin(x1))

plt.subplot(2, 1, 2)
plt.plot(x1, np.cos(x1));

plt.show()
```

#### **Working with Numpy Arrays**

```python
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green trinagles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
''' it will plot 3 graph together
    1st graph - 'r--' --> red & dashed line
    2nd graph - 'bs' --> blue & squared marker
    3rd graph - 'g^' --> green & Triangle marker
'''    
plt.show()
```

```python
fig = plt.figure()

x2 = np.linspace(0, 5, 10)
y2 = x2 ** 2

## [left margin, bottom margin, width margin, height margin]
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x2, y2, 'r')

axes.set_xlabel('x2')
axes.set_ylabel('y2')
axes.set_title('title');

plt.show()
```

#### **Stacked Bar Chart**

```python
A = [15. , 30. , 45. , 22. ]
B = [15. , 25. , 50. , 20. ]

z2 = range(4)

plt.bar(z2, A, color = 'b')
plt.bar(z2, B, color = 'r', bottom = A) # 'bottom = A' means bar 'B' start from 'A' bar end.

plt.show()
```

#### **Adding a grid**

```python
x15 = np.arange(1, 5)

# plot three graph together

plt.plot(x15, x15*1.5, 
         x15, x15*3.0, 
         x15, x15/3.0)

plt.grid(True)

plt.show()
```

#### **Handling axes**


```python
x15 = np.arange(1, 5)

plt.plot(x15, x15*1.5,
         x15, x15*3.0,
         x15, x15/3.0)

plt.axis()

plt.axis([0, 5, -1, 13])
plt.show()
```

```python
x15 = np.arange(1, 5)

plt.plot(x15, x15*1.5,
         x15, x15*3.0,
         x15, x15/3.0)
plt.xlim(1.0, 4.0)
plt.ylim(0.0, 12.0)

plt.show()
```

#### **Handling X and Y ticks**

```python
u = [5, 4, 9, 7, 8, 9, 6, 5, 7, 8]

plt.plot(u)

plt.xticks([2, 4, 6, 8, 10]) # x-axis ticks mark --> control which value appear on the x axis
plt.yticks([2, 4, 6, 8, 10]) # y-axis ticks mark --> control which value appear on the y axis

plt.show()
```

#### **Adding a legend**
