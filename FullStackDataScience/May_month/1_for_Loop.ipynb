```python
# Nested while loops

i = 1

while i <= 2:
    j = 0
    while j <= 2:
        print(i*j, end = " ")
        j +=1
    print()
    i+=1
```

```python
i = 1
while i <= 4:
    j = 0
    while j <= 3:
        print(i * j , end = " ")
        j += 1
    print()
    i += 1
```
