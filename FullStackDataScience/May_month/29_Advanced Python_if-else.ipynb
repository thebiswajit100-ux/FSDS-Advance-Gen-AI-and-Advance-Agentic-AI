#### Nested if

```python
x = 3
r = x % 2

if r == 0:
    print('Even number')

    if x > 5:
        print('greater number')

else:
    print('Odd Number')
```

```python
# nested if

x = 4
r = x % 2

if r == 0:
    print('Even number')

    if x>5:
        print('greater Number')

else:
    print('Odd number')
```

```python
# Nested if

x = 4
r = x % 2

if r == 0:
    print('Even number')

    if x > 5:
        print('greater number')

    else:
        print('lesser number')

else:
    print('Odd number')
```

#### loops

```python
i = 1   # initializing

while i <= 5:   # condition
    print('Data science')
    i = i + 1    # increment
```

```python
i = 1   # initializing
while i <= 5:   # condition
    print('data science:', i)
    i = i + 1   # increment
```

```python
i = 5    # initializing

while i>=1: # condition
    print('data science:', i)
    i = i -1  # decrement
```

```python
i = 5    # initiaziling
while i > 1:   # condition
    print('data science:' , i)
    i = i - 1  # decrement
```
