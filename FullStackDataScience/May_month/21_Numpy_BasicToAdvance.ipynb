###### **Speed of List Vs Numpy**

###### List

```python
# Element-wise addtion

a = [ i for i in range(10000000)]
b = [ i for i in range(10000000,20000000)]

c= []

import time

start = time.time()
for i in range(len(a)):
    c.append(a[i] + b[i])

print(time.time() - start)
```

###### Numpy

```python
import numpy as np

a = np.arange(10000000)
b = np.arange(10000000,20000000)

start = time.time()
c=a+b
print(time.time()-start)
```

> So, Numpy is faster than Normal Python Programming, we can see in above Example.
because Numpy uses C type array

#### Broadcasting (page 6)

> The term broadcasting describes how Numpy treats arrays with different shapes during
arithmetic operations

```python
# same shape
a = np.arange(6).reshape(2,3)
b = np.arange(6,12).reshape(2,3)

print(a,'\n') 
print(b,'\n')

print(a+b)
```

```python
# diff shape
a = np.arange(6).reshape(2,3)
b = np.arange(3).reshape(1,3)

print(a,'\n')
print(b,'\n')

print(a+b)
```

```python
a = np.arange(12).reshape(4,3)
b = np.arange(3)

print(a)  # 2 D
```

```python
print(b)
```

```python
print(a+b) # Arithmetic operation
```

```python
# Could not Broadcast

a  = np.arange(12).reshape(3,4)
b = np.arange(3)

print(a)
print(b)

print(a+b)
```

```python
a = np.arange(3).reshape(1,3)
b = np.arange(3).reshape(3,1)

print(a)
print(b)

print(a+b)
```

```python
a = np.arange(3).reshape(1,3)
b = np.arange(4).reshape(4,1)

print(a)
print(b)

print(a + b)
```

```python
a = np.array([1])
# shape -> (1,1) stretched to 2,2
b = np.arange(4).reshape(2,2)
# shape -> (2,2)

print(a)
print(b)

print(a+b)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return np.where((x**2 + y**2 < 1), 1.0, 0.0)

x = np.linspace(-5, 5, 500)
y = np.linspace(-5, 5, 500)
xv, yv = np.meshgrid(x,y)
rectangular_mask = f(xv, yv)

plt.pcolormesh(xv, yv, rectangular_mask, shading='auto')
plt.colorbar()
plt.grid()
plt.show()
```
