Python Operators and Expression Answers

```python
# Arithmetic operators
a = 5
b = 2
sum_result = a + b
difference_result = a - b
product_result = a * b
division_result = a / b

# comparision operator
is_equal = a == b
is_not_equal = a != b
is_greater_than = a > b
is_less_than = a < b

# Logical operator
logical_and = (a > 0) and (b > 0)
logical_or = (a > 0) or (b > 0)
logical_not = not (a > 0)

print("Arithmetic Operator:")
print("sum:", sum_result)
print("Difference:", difference_result)
print("Product:", product_result)
print("Division:", division_result)

print("\nComparision Operators:")
print("Is Equal:", is_equal)
print("Is NOt Equla:", is_not_equal)
print("Is Greater Than:", is_greater_than)
print("Is Less Than:", is_less_than)

print("\nLogical Operators:")
print("Logical AND:", logical_and)
print("Logical OR:", logical_or)
print("Logical NOt:", logical_not)
```

2. Difference between Unary and binary operator

```python
# Unary operators operate on a single operand.
number = 10
result = -number
print("Unary Minus Operator:", result)

#unary plus (+) operator
result = +number
print("Unary Plus Operator:", result)

# Logical Not (!) operator
is_true = True
logical_not_result = not is_true
print("Logical Not Operator:", logical_not_result)
```

Binary Operator: operate on two operands.

```python
# Binary addition (+) operator
a = 5
b = 3
addition_result = a + b
print("Binary Addition Operator:", addition_result)

# Binary multiplication (*) operator
mutiplication_result = a * b
print("Binary Multiplication Operator:", mutiplication_result)

# Logical AND (&) Operator
bitwise_and_result = a & b
print("Bitwise AND Operator:", bitwise_and_result)
```
