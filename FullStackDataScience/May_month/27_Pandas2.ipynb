```python
import matplotlib.pyplot as plt # visualization
import seaborn as sns  # distribution visualiztion

%matplotlib inline
plt.rcParams['figure.figsize'] = 6,2 # 6 --> width, 2--> height of the graph

import warnings
warnings.filterwarnings('ignore') # os error ( don't display warning object)
```

```python
df['InternetUsers']
```

```python
# .distplot  --> it combines histogram(bar) and Density Curves(smooth line)

vis1 = sns.distplot(df['InternetUsers'])
plt.show()
```

```python
# distribution plot --> it only present histogram

vis2 = sns.displot(df['InternetUsers'])
plt.show()
```

```python
# displot --> create histogram and bins = divide data into 15 groups

vis2 = sns.displot(df['InternetUsers'], bins=15)
plt.show()
```

```python
# displot --> by default create histogram

vis2 = sns.displot(df['InternetUsers'])
plt.show()
```

```python
# Box plot using seaborn
vis3 = sns.boxplot(data = df, x='IncomeGroup', y='BirthRate')
plt.show()
```

```python
# .lmplot --> Linear Model Plot --> create scatter plot and regression line automatically

vis4 = sns.lmplot(data = df, x = 'InternetUsers', y = 'BirthRate')
plt.show()
```

```python
# fig_reg = false --> disable regression line
vis5 = sns.lmplot(data = df, x = 'InternetUsers', y = 'BirthRate', fit_reg = False)
plt.show()
```

```python
# Scatter plot --> Adds different colours based on IncomeGroup
vis6 = sns.lmplot(data = df, x = 'InternetUsers', y = 'BirthRate', fit_reg = False, hue = 'IncomeGroup')
plt.show()
```

```python
## Scatter Plot Regression Analysis
vis6 = sns.lmplot(data = df, x = 'InternetUsers', y = 'BirthRate', fit_reg = True, hue = 'IncomeGroup')
plt.show()
```
