#### If Statement

```python
age = 20
if age >= 18:
    print('Eligible to vote.')
```

#### Short Hand if

Short-hand if is used to write if statements in a single line. it is useful when only one statement needs to executed.

```python
age = 19
if age > 18: print('Eligible to Vote.')
```

#### If else Statement

```python
age = 10
if age <= 12:
    print('Travel for free.')
else:
    print('Pay for ticket.')
```

#### If-elif-else Statement

```python
age = 25

if age <= 12:
    print('Child.')
elif age <= 19:
    print('Teenage')
elif age <= 35:
    print('Young adult.')
else:
    print('Adult')
```

#### Nested if-else Statement

```python
age = 70
is_member = True

if age >= 60:
    if is_member:
        print('30% senior discount!')
    else:
        print('20% senior discount.')
else:
    print('Not eligible for a senior discount.')
```

#### Ternary Conditional Statement

Ternary conditional statement is a short way to write an if-else statement in a single line. It is used when choosing between two values based on a condition.

```python
age = 17
s = "Adult" if age >= 18 else "Minor"
print(s)
```

#### Match-Case Statement

Match-Case statement is used to compare a value against multiple patterns and execute the matching block of code. it is similar to the switch-case statement available in other programming languages.

```python
number = 2

match number:
    case 1:
        print("One")
    case 2 | 3:
        print("Two or Three")
    case _:
        print("Other number")
```

#### Taking Conditional User Inputs - Python

```python
age = int(input("Enter your age:"))

if age >= 18:
    print("Eligible")
else:
    print("Not Eligible")
```

#### Types of Conditional User Inputs

1. Using  if and else conditions

   Example:

```python
num = int(input("Enter a number: "))

if num > 0:
    print("Positive Number")
else:
    print("Not Positive Number")
```

2. Using if, elif and else Conditions

   Example:

```python
marks = int(input("Enter your marks:"))

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Grade D")
```

3. Using Nested if Conditions

   Example:

```python
username = input("Enter username: ")
password = input("Enter password: ")

if username == "admin":
    if password == "1234":
        print("Login successful.")
    else:
        print("Incorrect password.")
else:
    print("Unknown username")
```

4. Using while Loop for Repeated Input

   Example:

```python
password = ""
while password != "python123":
    password = input("Enter the password:")

print("Access granted!")
```

5. Using try-except for Valid Input Type

   Example:

```python
while True:
    try:
        age = int(input("Enter your age:"))
        break
    except ValueError:
        print("Invalid input! Please enter a number.")

print("Thank you! You entered:", age)
```
